Nettoyer les paroles

```python
import re
# Fonction pour supprimer les éléments entre crochets dans une chaîne
def remove_brackets(text):
    res = re.sub(r'\[.*?\]', '', text)
    res = re.sub(r'\(.*?\)', '', res)
    return res

def remove_numbers(text):
    return re.sub(r'\d+', '', text)

def remove_useless_text(text):
    res = re.sub(r'Contributors', '', text)
    res = re.sub(r'Contributor', '', res)
    return re.sub(r'Lyrics', '', res)

# Appliquer la fonction à la colonne "Lyrics" pour nettoyer les paroles
df['Cleaned_Lyrics'] = df['Lyrics'].apply(remove_brackets)
df['Cleaned_Lyrics'] = df['Cleaned_Lyrics'].apply(remove_numbers)
df['Cleaned_Lyrics'] = df['Cleaned_Lyrics'].apply(remove_useless_text)

df['Cleaned_Lyrics'].str.replace('\n', ' ')
```

```python
# Supprimer les lignes qui n'ont pas [Verse ou [Chorus dans la colonne Lyrics 

print("Shape before removing rows: ", df.shape)
df = df[df['Lyrics'].str.contains(r'\[Verse|\[Chorus')]
print("Shape after removing rows: ", df.shape)
```
