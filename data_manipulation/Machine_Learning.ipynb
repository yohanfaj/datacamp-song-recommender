```python
import pandas as pd

df = pd.read_csv('dataset.csv')
df.head()
```

```python
df
```

```python
# Supprimer les lignes qui ont la meme valeur dans la colonne Name et Artist

df.drop_duplicates(subset=['Name', 'Artist'],  keep='first', inplace=True)
df.shape
```

```python
# Drop the columns that are not required
unwanted_columns = ["Artist", "Album", "Lyrics", "Cleaned_Lyrics", "Cleaned_Album", "moderation_result", "sentiment_result", "magnitude", "Artist_Encoded", "Album_Encoded"]
df.drop(unwanted_columns, axis=1, inplace=True)
df.head()
```

```python
# Normalize the data

from sklearn.preprocessing import StandardScaler

# Sélectionnez les colonnes à normaliser
colonnes_a_normaliser = ['Popularity']

# Initialisez l'objet StandardScaler
scaler = StandardScaler()

# Normalisez les colonnes sélectionnées
df[colonnes_a_normaliser] = scaler.fit_transform(df[colonnes_a_normaliser])

# Maintenant, les colonnes sont normalisées dans la DataFrame df
df.head()
```

K-means

```python
from sklearn.cluster import KMeans
colonnes_analyse = ['Popularity', 'Toxic',
       'Insult', 'Profanity', 'Derogatory', 'Sexual', 'Death, Harm & Tragedy',
       'Violent', 'Firearms & Weapons', 'Public Safety', 'Health',
       'Religion & Belief', 'Illicit Drugs', 'War & Conflict', 'Politics',
       'Finance', 'Legal', 'score']

# Créez un DataFrame contenant uniquement les colonnes d'analyse
data_analyse = df[colonnes_analyse]
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Créez une liste vide pour stocker les valeurs d'inertie
inertie = []

# Nombre maximal de clusters à explorer
nombre_max_clusters = 50

# Exécutez K-Means pour différents nombres de clusters et enregistrez l'inertie
for k in range(1, nombre_max_clusters + 1):
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(data_analyse)
    inertie.append(kmeans.inertia_)

# Tracez le diagramme du coude
plt.figure(figsize=(8, 6))
plt.plot(range(1, nombre_max_clusters + 1), inertie, marker='o', linestyle='-')
plt.xlabel('Nombre de clusters')
plt.ylabel('Inertie')
plt.title('Diagramme du coude')
plt.grid(True)
plt.show()
```

```python
# Spécifiez le nombre de clusters souhaité
nombre_de_clusters = 20

# Initialisez le modèle K-Means
kmeans = KMeans(n_clusters=nombre_de_clusters, random_state=0)

# Ajustez le modèle aux données
kmeans.fit(data_analyse)

# Obtenez les étiquettes de cluster attribuées à chaque ligne de la DataFrame
labels = kmeans.labels_

# Ajoutez les étiquettes de cluster à la DataFrame
df['Cluster_Labels'] = labels
```

```python
# afficher le nombre de chansons par cluster

df['Cluster_Labels'].value_counts()
```

```python
# Afficher les chansons pour chaque cluster

for i in range (0, nombre_de_clusters):
    print("Cluster", i)
    print(df[df['Cluster_Labels'] == i]['Name'])
    print()
```
